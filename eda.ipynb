## Word Cloud of Job Titles
This section visualizes the most frequent words in the job titles using a word cloud.

```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Load the dataset
# Make sure the file name matches your dataset
# If your job title column has a different name, update 'job_title' below

df = pd.read_excel("potential-talents.xlsx")

# Try to automatically find the job title column
possible_columns = [
    "job_title",
    "title",
    "position",
    "role",
    "job",
    "designation",
    "job title",
]
job_title_column = None
for col in df.columns:
    if any(keyword in col.lower() for keyword in possible_columns):
        job_title_column = col
        break

# If not found, set manually
# job_title_column = 'your_column_name_here'

if job_title_column:
    job_titles = df[job_title_column].dropna().astype(str)
    text = " ".join(job_titles)
    wordcloud = WordCloud(
        width=1200,
        height=600,
        background_color="white",
        colormap="viridis",
        stopwords={"and", "or", "the", "a", "an", "in", "at", "of", "for", "with"},
    ).generate(text)
    plt.figure(figsize=(15, 8))
    plt.imshow(wordcloud, interpolation="bilinear")
    plt.axis("off")
    plt.title("Word Cloud of Job Titles", fontsize=20)
    plt.show()
else:
    print("Job title column not found. Please specify it manually.")
```
